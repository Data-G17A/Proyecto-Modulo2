```python
df_4['Sport'] = df_4['Sport'].replace({'auto racing (nascar)':'auto racing'})
df_4['Sport'] = df_4['Sport'].replace({'f1 motorsports':'auto racing'})
df_4['Sport'] = df_4['Sport'].replace({'f1 racing':'auto racing'})
df_4['Sport'] = df_4['Sport'].replace({'nascar':'auto racing'})

df_4['Sport'] = df_4['Sport'].replace({'nfl':'american football'})
df_4['Sport'] = df_4['Sport'].replace({'american football / baseball':'american football'})

df_4['Sport'] = df_4['Sport'].replace({'nba':'basketball'})

df_4['Sport'] = df_4['Sport'].replace({'ice hockey':'hockey'})

df_4.pivot_table(columns=['Sport'], aggfunc='size')
```

7. Atleta con mayores ganancias por deporte por dÃ©cada

```python
df_4
```

```python
df_1990_2000=df_4[(df_4["Year"]>=1990) & (df_4["Year"]<=2000)]
df_2000_2010=df_4[(df_4["Year"]>2000) & (df_4["Year"]<=2010)]
df_2010_2020=df_4[(df_4["Year"]>2010) & (df_4["Year"]<=2020)]
```

```python
Total_earnings_1dec=df_1990_2000.groupby(["earnings ($ million)",'Name'])['Sport']

print(Total_earnings_1dec.sort_values(ascending=False))
```

```python
Total_earnings_2dec=df_2000_2010.groupby(['Name','Sport'])["earnings ($ million)"].sum()
print(Total_earnings_2dec.sort_values(ascending=False))
```

```python
Total_earnings_3dec=df_2010_2020.groupby(['Name','Sport'])["earnings ($ million)"].sum()
print(Total_earnings_3dec.sort_values(ascending=False))
```
